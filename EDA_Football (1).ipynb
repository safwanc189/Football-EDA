```python
df1['Height'] = df1['Height'].str.replace('cm','')

df1['Weight'] = df1['Weight'].str.replace('kg','')

df1['Contract'] = df1['Contract'].str.replace('~','-')

df1.rename(columns={'Height':'Height in cm','Weight':'Weight in kg'},inplace =True)
```

```python
df1['Height in cm'].unique()
```

```python
def height_to_cm(height):
    if "'" in height and '"' in height:
        feet, inches = height.split("'")
        inches = inches.replace('"', '')
        feet = int(feet)
        inches = int(inches)
        return feet * 30.48 + inches * 2.54
    else:
        return float(height)
```

```python
df1['Height in cm'] = df1['Height in cm'].apply(height_to_cm)
```

```python
df1['Height in cm'] = df1['Height in cm'].astype(int)
```

```python
df1['Height in cm'].unique()
```

```python
df1['Weight in kg'].unique()
```

```python
def weight_to_kg(weight):
    if 'lbs' in weight:
        lbs = weight.replace('lbs', '')
        lbs = float(lbs)
        return lbs * 0.453592
    else:
        return float(weight)
```

```python
df1['Weight in kg'] = df1['Weight in kg'].apply(weight_to_kg)
```

```python
df1['Weight in kg'] = df1['Weight in kg'].astype(int)
```

## **df2**

```python
df2 = df.iloc[:,15:26]
df2.head()
```

```python
df2['Value'] = df2['Value'].str.replace("€", "")
```

```python
df2
```

```python
import numpy as np
```

```python
def convert_EM(x):
    if isinstance(x, str):
        if "K" in x:
            return float(x.replace("K", "")) / 1000
        elif "M" in x:
            return float(x.replace("M", ""))
    return float(x)

df2['Value'] = df2['Value'].apply(convert_EM)
```

```python
df2.rename(columns={'Value':'Values in Euro Million'},inplace =True)
```

```python
df2
```

```python
def convert_wage(x):
    if isinstance(x, str):
        x = x.replace("€", "")
        if "K" in x:
            return float(x.replace('K', '')) * 1000
        else:
            return float(x)
    return float(x)

df2['Wage'] = df2['Wage'].apply(convert_wage)
```

```python
df2.rename(columns={'Wage':'Wage in Euros'},inplace =True)
```

```python
df2
```

```python
df2['Release Clause'] = df2['Release Clause'].str.replace("€", "")
df2['Release Clause'] = df2['Release Clause'].apply(convert_EM)
df2.rename(columns={'Release Clause':'Release Clause in Euro Million'},inplace =True)
```

```python
df2
```

```python
sns.lineplot(data=df2,x='Wage in Euros',y='Release Clause in Euro Million',color='green')
plt.title('Wage vs Release Clause')
plt.show()
```

```python
df2['Wage in Euros'] = df2['Wage in Euros'].astype(int)
```
